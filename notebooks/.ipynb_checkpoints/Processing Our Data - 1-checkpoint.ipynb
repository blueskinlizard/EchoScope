# Welcome to EchoScope!

### Or more specifically, the data processing notebook.
This notebook serves as the training data preparation pipeline for EchoScope, a deep learning‚Äìpowered radio signal modulation classifier. While the end goal of EchoScope is to accept raw RF .wav files from users via a web interface and predict their modulation type, this notebook focuses specifically on preprocessing and augmenting training data to ensure the underlying model is robust, generalizable, and performant.

### What is EchoScope's model architecture? (IMPORTANT)
EchoScope employs a dual-branch neural architecture designed to extract and fuse complementary insights from both raw and transformed signal representations. The model is tailored specifically for classifying radio modulation types from noisy, real-world radio data.

#### Branch 1: IQ Time-Series Processing
The first branch directly consumes IQ (in-phase and quadrature) time-series data, preserving the temporal structure of the raw radio signal. This data is passed through a Bidirectional LSTM (Long Short-Term Memory) network, which captures both forward and backward temporal dependencies, allowing the model to understand modulation patterns that evolve over time.

#### Branch 2: Spectrogram-Based Processing
In parallel, the same IQ data is transformed into a spectrogram, representing the frequency content over time. This visual-like representation is then processed using a Transformer-based Patch Encoder inspired by the Vision Transformer (ViT) paradigm. The spectrogram is divided into non-overlapping patches, each of which is embedded and passed through a multi-layer Transformer Encoder to capture long-range spatial-frequency dependencies. Spectrogram dimensionality is (2, 17, 29)

#### Fusion branch
EchoScope uses a cross-attention-based fusion to integrate features from both branches:

The LSTM (IQ) and Transformer (spectrogram) branches each output learned feature vectors.

A multi-head attention layer lets the IQ features attend to the spectrogram features, refining the time-series representation with frequency-domain context.

The attended IQ features are residually connected back to the original.

Both feature vectors are then concatenated and passed through a fully connected fusion network, producing the final prediction.

Auxiliary classifiers on each branch provide additional outputs for interpretability and regularization.

### What is this graph?
This graph shows the IQ of an example. 

The I/Q (In-phase and Quadrature, not intelligence quotient) components represent the real and imaginary parts of a complex baseband signal.

Together, they capture both amplitude and phase information of a radio waveform over time. Plotting them helps visualize modulation patterns and assess signal characteristics (but we won't be spamming plots in this notebook, statistical data will work just fine for analysis).


```python
def process_iq_data(samples):
    normalized = np.empty_like(samples)
    for i in range(samples.shape[0]):
        sample = samples[i]
        for ch in range(2):
            mean = sample[ch].mean()
            std = sample[ch].std()
            normalized[i][ch] = (sample[ch] - mean) / (std + 1e-10)
    return normalized
```

```python
for key in mods_snrs:
    samples = radio_ml_2016[key]
    radio_ml_2016[key] = process_iq_data(samples)
```

Here, each IQ sample is channel-wise normalized to have zero mean and unit variance. For both the I and Q channels, the mean and standard deviation are computed independently.

```python
def generate_spectrogram(iq_sample, fs=1.0, nperseg=32, noverlap=24, 
                          noise_std=0.01, shift_max=2,
                         pad_to=256):
    iq_sample = iq_sample / (np.linalg.norm(iq_sample, axis=1, keepdims=True) + 1e-10)
    
    if pad_to > iq_sample.shape[1]:
        pad_width = pad_to - iq_sample.shape[1]
        iq_sample = np.pad(iq_sample, ((0, 0), (0, pad_width)), mode='reflect')

    spec_channels = []
    for ch in range(iq_sample.shape[0]):
        f, t, Sxx = spectrogram(iq_sample[ch], fs=fs, nperseg=nperseg, noverlap=noverlap)
        Sxx = np.abs(Sxx)
        Sxx_db = 10 * np.log10(Sxx + 1e-10)
        spec_channels.append(Sxx_db)

    spectrogram_img = np.stack(spec_channels, axis=0)
    
    return spectrogram_img
```

```python
spectrograms_dict = {}

for key in tqdm(mods_snrs, desc="Generating spectrograms"):
    samples = radio_ml_2016[key]
    spectrograms = []
    for i in range(samples.shape[0]):
        spec_img = generate_spectrogram(samples[i], )
        spectrograms.append(spec_img)
    spectrograms = np.array(spectrograms)
    spectrograms_dict[key] = spectrograms 
```

```python
all_specs = []
for key in spectrograms_dict:
    all_specs.extend(spectrograms_dict[key].copy())

all_specs = np.stack(all_specs) 
global_mean = all_specs.mean()
global_std = all_specs.std()

for key in spectrograms_dict:
    spectrograms_dict[key] = (spectrograms_dict[key] - global_mean) / (global_std + 1e-9)

np.save("global_mean.npy", global_mean)
np.save("global_std.npy", global_std)

print("Post-standardization stats:")
print("Mean:", global_mean)
print("Std:", global_std)
```

### What does this code do?

This function takes our IQ signal list and computes a spectrogram for each channel.

A spectrogram shows how the signal‚Äôs frequency content changes over time. For each channel, it calculates the spectrogram, converts the power values to a decibel scale for easier interpretation, and then collects all these spectrograms together into one 3D array. The result is an array where each channel has its own frequency-vs-time spectrogram image, useful for analyzing the signal‚Äôs characteristics across different channels.

```python
def print_spectrogram_stats(spectrograms_dict, num_keys=20):
    for key, specs in islice(spectrograms_dict.items(), num_keys):
        num_samples = specs.shape[0]
        sample_shape = specs.shape[1:]
        min_val = specs.min()
        max_val = specs.max()
        mean_val = specs.mean()
        nan_count = np.isnan(specs).sum()
        inf_count = np.isinf(specs).sum()
        total_vals = specs.size

        print(f"Key: {key}")
        print(f" Number of samples: {num_samples}")
        print(f" Spectrogram shape per sample: {sample_shape}")
        print(f" Min value: {min_val:.4f}")
        print(f" Max value: {max_val:.4f}")
        print(f" Mean value: {mean_val:.4f}")
        print(f" NaNs present: {nan_count > 0} ({nan_count} / {total_vals}, {100 * nan_count / total_vals:.5f}%)")
        print(f" Infinite values present: {inf_count > 0} ({inf_count} / {total_vals}, {100 * inf_count / total_vals:.5f}%)")
        print("-" * 40)

print_spectrogram_stats(spectrograms_dict, num_keys=20)
```

## Woah, isn't (17, 29) kinda coarse for a spectrogram?

Yeah, it does seem a bit rough at first, but surprisingly, it works really well?

While playing around with the Spectrogram branch in EchoScope, I tested a bunch of different shapes. Turns out, spectrograms with a height of 17 actually gave us the best accuracy. Weirdly enough, when we bumped it up to something like 65, the performance dropped (a LOT!). Even 33 didn‚Äôt offer much of a gain.

So why (17, 29)? Honestly, speed. It gave us fast training times and solid accuracy. It‚Äôs not the most detailed spectrogram in the world, but it gets the job done really well without slowing everything down.

When we increased the dimensions of (17, 29) to something like (17, 129), epoch training time would nearly double, and accuracy would only be improved by 1-2%.
